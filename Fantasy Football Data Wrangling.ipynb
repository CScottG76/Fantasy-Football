rbs14 df has been appropriately split between weekly dfs.  Tidying is necessary:  The w1 df differs from subsequent dfs with an additional column.  The column names still indicate repetition (i.e. column, column.1, etc) and will have to match for concatination.

```python
#Removing additional column in first week's df.
w1 = w1.iloc[:,0:13]
```

```python
#Creating iterable list of dfs & list of column names.  Loop created to match column names over dfs. 
df_list = [w1, w2, w3, w4, w5, w6, w7, w8, w9, w10, w11, w12, w13, w14, w15, w16, w17]
col_names = ['Player','Team','Wk','Run Att','Run Yd','Run TD', 'Target','Rec','Rec Yd','Rec TD','Opp','Points','Game played']


for df in df_list:
    df.columns = col_names
```

```python
#Concatinating weekly dfs into single df.  
df14 = pd.concat(df_list)
df14.head()
```

Two values in column indicated if Player participated in the game in the given week.  Zero's indicate play did not play and, therefore, rows not needed in df.

```python
#Removing rows where player did not play.
df14 = df14[df14['Game played'] == 1]
```
